### Observations
* The average tempertures are highest for countries near the equator. 
* Warm air can hold more water therefore countries with higher temperatures will be more humid.  
* Wind has little impact on the temperature. 
* Cloudiness can cause a greenhouse effect which can increase temperatures regardless of location. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information, setup counters and print header for display data.

url = "http://api.openweathermap.org/data/2.5/weather?"
query_cnt = 0
query_url = url + "appid=" + api_key + "&q="
unit_url = '&units=imperial'

weather = []

# Print the header for the display.

print("Beginning Data Retrieval")
print("--------------------------------")

# Loop through all of the cities stored from the lat/long calls and gather the weather data

for city in cities:
    try:
        # Build query URL and increment the query counter
        query_cnt += 1
    
        # Print the query record count and city name followed by the url used to gather the weather data. 
        print("Processing Record " + str(query_cnt) + " | " + city) 
        print(query_url)
        
        response = requests.get(query_url + city + unit_url).json()
        
        # Get the specific column data
        lat = (response['coord']['lat'])
        long = (response['coord']['lon'])
        temp =(response['main']['temp'])
        humidity = (response['main']['humidity'])
        cloudiness = (response['clouds']['all'])
        wind = (response['wind']['speed'])
        country = (response['sys']['country'])
        
        # Add the returned data to the weather list
        weather.append({"City": city,
                    "Country": country,
                    "Lat": lat, 
                    "Long": long,
                    "Temp": temp,
                    "Humidity": humidity,
                    "cloudiness": cloudiness,
                    "wind": wind})
    except:
        # City not found error
        print(city + " not found ")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df = pd.DataFrame(weather)
weather_df

# Export the new CSV
weather_df.to_csv("Weather_New.csv", index=False)

# View the DataFrame
weather_df.head()
```

#### Latitude vs. Temperature Plot

```python
# Calculate the average temperature and average latitude by country. 
avg_lat = weather_df.groupby(['Country'])['Lat'].mean()
avg_lat

avg_temp = weather_df.groupby(['Country'])['Temp'].mean()
avg_temp

avg_humidity = weather_df.groupby(['Country'])['Humidity'].mean()
avg_humidity

avg_cloud = weather_df.groupby(['Country'])['cloudiness'].mean()
avg_cloud

avg_wind = weather_df.groupby(['Country'])['wind'].mean()
avg_wind

# Save the calculations into a summary data frame
avg_lat_df = pd.DataFrame({"avg_lat": avg_lat,
                           "avg_temp": avg_temp,
                           "avg_humidity": avg_humidity,
                           "avg_cloud": avg_cloud,
                           "avg_wind": avg_wind})
avg_lat_df.head()
```

```python
# The maximum x value for the chart will be 100
x_limit = 100

# List of values from 0 to 100 each value being 5 greater than the last
x_axis = np.arange(0, x_limit, 5)

# Build scatter plot
avg_lat_df.plot(kind='scatter', x='avg_lat', y='avg_temp', marker="o", c="green", edgecolors="black", s=avg_lat_df.avg_temp)

# Incorporate the other graph properties
plt.title("Average Temp by Latitude")
plt.xlabel("Latitude")
plt.ylabel("Average Temp")
plt.grid()
```
